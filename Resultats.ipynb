Affichage des résultats avec un heatmap

```python
n_row = 3
n_col = 2
fig, axs = plt.subplots(n_row, n_col, figsize=(10,10))
count = 0
labels = ['Meningioma', 'Glioma', 'Pituitary']
for key, value in experiments.items():
    hm = sns.heatmap(value, annot=True, ax=axs[int(count/n_col), count%n_col], xticklabels=labels, yticklabels=labels)
    hm.set_title(key)
    hm.set_xlabel('Prédictions')
    hm.set_ylabel('Étiquette')
    count += 1

# Mise en forme
pad = 10 # in points
rows = ['Original', 'Égalisé', 'Augmenté']

for ax, row in zip(axs[:,0], rows):
    ax.annotate(row, xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - pad, 0),
                xycoords=ax.yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')
fig.tight_layout()
fig.savefig("figures/matrices_confusion.png")
```
